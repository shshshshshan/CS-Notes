# Simple Regression Process

Simple regression done using Python

```python
# Importing necessary modules
import numpy as np
import matplotlib.pyplot as plt
```

## Dataset

The sample dataset is a relationship between a house's size in square feet and its corresponding price in $1000s.

```python
# Dataset
dataset = [
    (2104, 400),
    (1416, 232),
    (1534, 315),
    (852, 178),
    (600, 100)
]
```

## Regression Process

### Simple regression formula:
$$
  y = \beta_0 + \beta_1 x
$$

where: <br>
  * $y_{(i)}$ - Estimated (or predicted) $y$ value for $i$-th observation.
  * $\beta_0$ - Estimate of the regression intercept
  * $\beta_1$ - Estimate of the regression slope
  * $x_{(i)}$ - Value of $x$ for $i$-th observation.

---

### Steps in taking the simple regression
  * Get the mean of the x- and y-values of the dataset. These are denoted as $\bar{x}$ and $\bar{y}$.
  * Calculate $\beta_1$ or the intercept. This is defined as the function: 
  
  $$
  \beta_1 = \frac { \sum (x_i - \bar{x}) (y_i - \bar{y}) } {\sum (x_i - \bar{x})^2}
  $$ 
  
  <center>or</center> 
  
  $$
  \beta_1 = \frac { n\sum {x_i y_i}- \sum{x_i} \sum{y_i} } {n\sum {x}^2 - (\sum{x_i})^2}
  $$

  * Calculate $\beta_0$ or the slope of the regression line. This is defined as the function: 
  
  $$
  \beta_0 = \bar{y} - \beta_1 \bar{x}
  $$ 
  
  <center>or</center> 
  
  $$
  \beta_0 = \frac{\sum{y_i} - \beta_1 \sum{x_i}} {n}
  $$

  > Where $n$ is the length of the dataset.

  * Plot the regression line by plugging sample values of $x$ to the regression formula and substituting the acquired values.

---

## Applying the process

### Step 1. Getting the mean of the $x$ and $y$

```python
# Extracting x and y values from the dataset
x = np.array([data[0] for data in dataset])
y = np.array([data[1] for data in dataset])

# Calculating means
xmean = np.mean(x)
ymean = np.mean(y)
```

### Step 2. Solving $\beta_1$

```python
# Calculating beta_1 and beta_0 using the formulas for simple linear regression
numerator = np.sum((x - xmean) * (y - ymean))
denominator = np.sum((x - xmean) ** 2)
beta_1 = numerator / denominator
```

### Step 3. Solving $\beta_0$

```python
beta_0 = ymean - beta_1 * xmean
```

```python
# The values of the variables
print(f'Beta_0: {beta_0}')
print(f'Beta_1: {beta_1}')
```

### Step 4. Plotting the Regression Line

```python
# Plotting the data points
plt.scatter(x, y, label='Data points')
plt.xlabel('House Size (sqft)')
plt.ylabel('Price ($1000s)')

# Plotting the regression line
x_pred = np.linspace(x.min(), x.max(), 50)
y_pred = beta_0 + beta_1 * x_pred
plt.plot(x_pred, y_pred, color='red', label='Regression Line')

# Displaying the plot with legend
plt.legend()
plt.show()
```
